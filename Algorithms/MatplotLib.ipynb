```python
x=np.arange(0,5,0.5)
x1=x**3

y=np.arange(5,0,-0.5)       
y1=y**3+15

plt.plot(x,x1,color='black',marker='o',linewidth=4,label='for x') # u can specificaly change the color and marker types
plt.plot(y,y1,color='blue',marker='p',linewidth=3,label='for y')  # u can change the line width also

plt.legend()                                                      # this help to put label inside the graph

plt.ylabel('y**3-20')                                # u can give the label to particular axis
plt.xlabel('x**3')

plt.title('FIRST GRAPH')                             # u can give the title

plt.grid()                                           # gives u grid

plt.axis([0,5,0,100])                                # u can change the axis

plt.text(2,60,'MIDDLE',fontsize=20)                  # to add a text in bettwen the graph 

plt.show()    
```

```python
# Pie Graphs

sizes=[3,4,6,5]                                    # to give the % size and partitions

plt.axis("equal")                                  # to make graph eual form all side

labels=['blue','orange','green','yellow']          # to give the labesl to sections

colors=['blue','orange','green','yellow']          # to fill diff colors

explode=[0.1,0.2,0,0]                              # to highlight a section , (couterclock) used to rotate anticlock wise

plt.title("Pie Graph",fontsize=30)                 # to give the title , (autopct) used for give the data in sections

plt.pie(sizes,colors=colors,labels=labels,explode=explode,autopct='%.1f%%',startangle=30,counterclock=False)

plt.show()
                                              
```
