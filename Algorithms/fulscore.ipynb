```python
import pandas as pd
import numpy as np
t=pd.read_csv('LOG_titanic_x_test.csv',delimiter=',')
nt=t.copy()
nt.Age.fillna(t.Age.mean(),inplace=True)
del nt['Name']
del nt['Ticket']
#del nt['Fare']
del nt['Cabin']
#del nt['Embarked']
#del nt['SibSp']
#del nt['Parch']
def f(s):
    if s=='male':
        return 1
    else :
        return 0

nt['Gender']=nt.Sex.apply(f)
del nt['Sex']

def z(s):
    if s=='C':
        return 0
    elif s=='S' :
        return 1
    else :
        return 3
    

nt['embark']=nt.Embarked.apply(z)
del nt['Embarked']
nt['age2']=(nt['Age']-nt['Fare'])/2


j=pd.read_csv('LOG_titanic_x_y_train.csv',delimiter=',')
Ns=j[j.Survived==0]
Ys=j[j.Survived==1]
Ns.Age.fillna(Ns.Age.mean(),inplace=True)
Ys.Age.fillna(Ys.Age.mean(),inplace=True)
newt=pd.concat([Ns,Ys])
del newt['Name']
del newt['Ticket']
#del newt['Fare']
del newt['Cabin']
#del newt['Embarked']
#del newt['SibSp']
#del newt['Parch']
def f(s):
    if s=='male':
        return 1
    elif s=='female' :
        return 0
newt['Gender']=t.Sex.apply(f)
del newt['Sex']

def z(s):
    if s=='C':
        return 0
    elif s=='S' :
        return 1
    else :
        return 3
newt['embark']=newt.Embarked.apply(z)
del newt['Embarked']
newt['age2']=(newt['Age']-newt['Fare'])/2
x=newt['Survived']
del newt['Survived']
newt['survived']=x
newt.Gender.fillna(newt.Gender.mean(),inplace=True)
data=np.array(newt)


x_test=np.array(nt)
data=np.array(newt)
x_train=data[:,0:-1]
y_train=data[:,-1]

from sklearn.linear_model import LogisticRegression
cf=LogisticRegression()
cf.fit(x_train,y_train)

y_pred=cf.predict(x_test) 
y=np.array(y_pred)

y_test_pred=np.around(y,decimals=5)
np.savetxt("LOGscore.csv",y_pred)
cf.score(x_test,y_test_pred)
y_pred
```
